````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
mov_url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(mov_url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
movies=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
print(movies.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
movie_item = movies.find_all('div', class_='title-list-grid__item')

# Movie url's extraction
movies_url = [item.find('a')['href'] for item in movie_item][:100]

# Print the movie links
for i, movie_url in enumerate(movies_url, start=1):
    print(f"{i}. {movie_url}")
````

## **Scrapping Movie Title**

````python
# Write Your Code here
movie_title = [item.find('img')['alt']for item in movie_item ][:100]

# Print movies title
for i, title in enumerate(movie_title, start=1):
    print(f"{i}. {title}")
````

## **Scrapping release Year**

````python
# Write Your Code here
main_url = "https://www.justwatch.com"
movies_release_years = []

for movie_url in movies_url:
  URL = main_url + movie_url

  response = requests.get(URL)
  soupx = BeautifulSoup(response.text, 'html.parser')

  release_year_element = soupx.find('div', class_ = 'title-block')
  release_year_str = release_year_element.find_next('span').text.strip()
  release_year = int(release_year_str.strip('()'))
  movies_release_years.append(release_year)

# Print movies release years
for i, years in enumerate(movies_release_years, start=1):
    print(f"{i}. {years}")
````

## **Scrapping Genres**

````python
# Write Your Code here
main_url = "https://www.justwatch.com"
movies_genres = []

for movie_url in movies_url:
  URL = main_url + movie_url

  response = requests.get(URL)
  soupx = BeautifulSoup(response.text, 'html.parser')

  movie_genre_element = soupx.find('h3', string= 'Genres')
  genre = movie_genre_element.find_next('div', class_= 'detail-infos__value').text.strip()

  movies_genres.append(genre)

# Print movies genres
for i, gen in enumerate(movies_genres, start=1):
  print(f"{i}. {gen}")
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
main_url = "https://www.justwatch.com"
movies_IMDB_rating = []

for movie_url in movies_url:
  URL = main_url + movie_url

  response = requests.get(URL)
  soupx = BeautifulSoup(response.text, 'html.parser')

  movie_rating_element = soupx.find('img', {'alt': 'IMDB'})
  if movie_rating_element:
      imdb_rating = movie_rating_element.find_next('span').text.strip()
      movies_IMDB_rating.append(float(imdb_rating.split()[0]))
  else:
      movies_IMDB_rating.append(None)

# Print movies IMDB rating
for i, rating in enumerate(movies_IMDB_rating, start=1):
  print(f"{i}. {rating}")
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
main_url = "https://www.justwatch.com"
movies_runtime = []

for movie_url in movies_url:
  URL = main_url + movie_url

  response = requests.get(URL)
  soupx = BeautifulSoup(response.text, 'html.parser')

  movie_runtime_element = soupx.find('h3', string= 'Runtime')
  runtime = movie_runtime_element.find_next('div', class_= 'detail-infos__value').text.strip()
  movies_runtime.append(runtime)

# Print movies runtime
for i, runtime in enumerate(movies_runtime, start=1):
  print(f"{i}. {runtime}")
````

## **Scrapping Age Rating**

````python
# Write Your Code here
main_url = "https://www.justwatch.com"
movies_age_rating = []

for movie_url in movies_url:
  URL = main_url + movie_url

  response = requests.get(URL)
  soupx = BeautifulSoup(response.text, 'html.parser')

  movie_age_rating_element = soupx.find('h3', string = 'Age rating')
  if movie_age_rating_element:

        age_rating_value = movie_age_rating_element.find_next('div', class_='detail-infos__value').text.strip()
        movies_age_rating.append(age_rating_value)
  else:
        movies_age_rating.append(None)

# Print movies runtime
for i, a_rating in enumerate(movies_age_rating, start=1):
  print(f"{i}. {a_rating}")
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
main_url = "https://www.justwatch.com"
movies_origin_country = []

for movie_url in movies_url:
  URL = main_url + movie_url

  response = requests.get(URL)
  soupx = BeautifulSoup(response.text, 'html.parser')

  movie_origin_country_element = soupx.find('h3', string=lambda text: text and "Production country" in text.strip())
  if movie_origin_country_element:

      country = movie_origin_country_element.find_next('div', class_= 'detail-infos__value').text.strip()
      movies_origin_country.append(country)
  else:
     movies_origin_country.append('Not Available')
# Print movies production country
for i, country in enumerate(movies_origin_country, start=1):
  print(f"{i}. {country}")
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
main_url = "https://www.justwatch.com"
movies_streaming_platform = []

for movie_url in movies_url:
  URL = main_url + movie_url

  response = requests.get(URL)
  soupx = BeautifulSoup(response.text, 'html.parser')

  movie_streaming_platform_element = soupx.find('div', class_='buybox-row stream')
  if movie_streaming_platform_element:

        platform_links = movie_streaming_platform_element.find('a', class_='offer')
        platform_names = platform_links.find('img')['alt']
        movies_streaming_platform.append(platform_names)
  else:
        movies_streaming_platform.append(None)

# Print movies streaming platform
for i, platform_names in enumerate(movies_streaming_platform, start=1):
  print(f"{i}. {platform_names}")
````
